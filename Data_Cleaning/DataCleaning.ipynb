```python
cols_with_missing = ['Teacher_Quality', 'Parental_Education_Level', 'Distance_from_Home']

for col in cols_with_missing:
    mode_val = df[col].mode()[0]
    df[col] = df[col].fillna(mode_val)
    print(f'{col} → filled with "{mode_val}"')
```

```python
df.isnull().sum()
```

```python
bool_cols = ['Extracurricular_Activities', 'Internet_Access', 'Learning_Disabilities']
for col in bool_cols:
    df[col] = df[col].map({'Yes': True, 'No': False})
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
num_cols = ['Hours_Studied', 'Attendance', 'Sleep_Hours', 'Previous_Scores','Tutoring_Sessions', 'Physical_Activity', 'Exam_Score']
```

```python
df[num_cols].plot(kind='box',subplots=True, layout=(2,4), figsize=(16,8))
```

```python
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)]
```

```python
df['Exam_Score'] = df['Exam_Score'].clip(upper=100)
print('Max Exam_Score is now:', df['Exam_Score'].max())
```

```python
df['Attendance'] = df['Attendance'].clip(lower=0, upper=100)
```

```python
str_cols = df.select_dtypes(include='object').columns
for col in str_cols:
    df[col] = df[col].str.strip().str.title()
for col in str_cols:
    print(f'{col}: {df[col].unique()}')
```

```python
from sklearn.preprocessing import LabelEncoder
df_encoded = df.copy()
le = LabelEncoder()
ordinal_cols = ['Parental_Involvement','Access_to_Resources','Motivation_Level','Family_Income','Teacher_Quality','Parental_Education_Level','Distance_from_Home','Peer_Influence'
]
for col in ordinal_cols:
    df_encoded[col] = le.fit_transform(df_encoded[col])
    print(f'{col} → {df_encoded[col].unique()}')
```
